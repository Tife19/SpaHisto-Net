```python
from google.colab import drive
import pandas as pd
from huggingface_hub import hf_hub_download
from pathlib import Path
import os

# ========================= 1. Mount Drive =========================
drive.mount('/content/drive')

# ========================= 2. Load Metadata =========================
meta_path = "/content/drive/My Drive/MSC Project/SpaHisto-Net1/data/raw/STimage-1k4M/meta_all_gene.csv"   # ← Change if your path is different
df = pd.read_csv(meta_path)

# Filter breast samples (both cancer and normal)
breast_df = df[df['tissue'] == 'breast'].copy()

print(f"Total breast samples found: {len(breast_df)}")
print(f"Cancer samples: {len(breast_df[breast_df['involve_cancer'] == True])}")
print(f"Normal samples: {len(breast_df[breast_df['involve_cancer'] == False])}")

# ========================= 3. Set Output Folder =========================
output_base = Path("/content/drive/My Drive/MSC Project/SpaHisto-Net1/data/raw/STimage-1k4M")
output_base.mkdir(parents=True, exist_ok=True)

# ========================= 4. Download Function =========================
def download_sample(row):
    slide = row['slide']
    tech = row['tech']                    # ST, Visium, or VisiumHD
    save_dir = output_base / slide
    save_dir.mkdir(parents=True, exist_ok=True)

    try:
        # Download image
        hf_hub_download(repo_id="jiawennnn/STimage-1K4M",
                        filename=f"{tech}/image/{slide}.png",
                        local_dir=save_dir, repo_type="dataset")

        # Download coordinates
        hf_hub_download(repo_id="jiawennnn/STimage-1K4M",
                        filename=f"{tech}/coord/{slide}_coord.csv",
                        local_dir=save_dir, repo_type="dataset")

        # Download gene expression
        hf_hub_download(repo_id="jiawennnn/STimage-1K4M",
                        filename=f"{tech}/gene_exp/{slide}_count.csv",
                        local_dir=save_dir, repo_type="dataset")

        # Annotation
        try:
            hf_hub_download(repo_id="jiawennnn/STimage-1K4M",
                            filename=f"annotation/{slide}_anno.csv",
                            local_dir=save_dir, repo_type="dataset")
            print(f"✅ Downloaded {slide} (with annotation)")
        except:
            print(f"✅ Downloaded {slide} (no annotation available)")


    except Exception as e:
        print(f"❌ Failed {slide} ({tech}): {e}")

# ========================= 5. Download All Breast Samples =========================
for _, row in breast_df.iterrows():
    download_sample(row)

print("\n🎉 Download completed! All files saved in:")
print(output_base)
```

```python
import pandas as pd

raw_data_dir = PROJECT_ROOT / "data" / "raw" / "STimage-1k4M"

print("=" * 80)
print("📁 FULL STRUCTURE INSPECTION — All 205 Samples")
print("=" * 80)

if not raw_data_dir.exists():
    print(f"\n❌ ERROR: Folder not found → {raw_data_dir}")
else:
    sample_folders = sorted([f for f in raw_data_dir.iterdir() if f.is_dir()])
    print(f"\nTotal samples: {len(sample_folders)}")

    records = []
    for sample in sample_folders:
        files = list(sample.iterdir())
        file_names = [f.name for f in files]

        record = {
            "sample": sample.name,
            "num_files": len(files),
            "has_image": any(".png" in f.lower() for f in file_names),
            "has_coords": any("_coord.csv" in f.lower() for f in file_names),
            "has_counts": any("_count.csv" in f.lower() for f in file_names),
            "has_annotation": any("_anno.csv" in f.lower() for f in file_names),
            "files": ", ".join(sorted(file_names))
        }
        records.append(record)

    df = pd.DataFrame(records)

    print("\n" + "-" * 80)
    print("Summary of file presence across all samples:")
    print("-" * 80)
    print(df[["has_image", "has_coords", "has_counts", "has_annotation"]].sum())

    print("\n" + "-" * 80)
    print("Full table (all 205 samples):")
    print("-" * 80)

    pd.set_option("display.max_rows", 210)
    pd.set_option("display.max_colwidth", 120)
    display(df)

    print("\n" + "=" * 80)
    print("✅ Done. You can now scroll through the table above.")
    print("=" * 80)
```
